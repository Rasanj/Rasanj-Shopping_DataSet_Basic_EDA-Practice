## Univariate Analysis

### Task 1 : User age distribution 

```python
df['Age'].describe()
```

```python
df['Age'].plot(kind='hist',color='blue',alpha=0.5,title='Age Distribution',xlabel='Age', ylabel='Frequency',figsize=(5,4),bins=20)
plt.show()
```

### Grouping of user_ages 

```python
def age_group(x):
    if 0<x<=24:
            return "18 - 25"
    elif 24<x<=35:
        return "25-35"
    elif 35<x<=45:
        return "35-45"
    elif 45<x<=55:
        return "45-55"
    elif 55<x<=65:
        return "55-65"
    elif 65<x<=85:
        return "65-85"
    else :
        return "Other"

df['Age_group']=df['Age'].apply(age_group)
df.head()
```

### bar plot based on age group 

```python
age_distribution= df['Age_group'].value_counts()
print(age_distribution)
```

```python
age_distribution.plot(figsize=(5,4),kind='bar',xlabel='Age Group',ylabel='Frequesncy',alpha=0.6,color='green',title='Age Distribution by age group')
plt.show()
```

### Task 8 : the most common payment methods used by customers

```python
Most_revenue_generating_payment_method = df.groupby('Payment Method')['Purchased_Amount'].sum()
Most_revenue_generating_payment_method .plot(figsize=(5,4),kind='line',color=['skyblue'],marker='o',fontsize=7, grid=True,                             
                             ylabel='Revenue',title='Payment Methods Usage Frequency')
plt.show()
```
